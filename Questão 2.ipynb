O objetivo dessa questão é a Feature Engineering para modelo com varíavel dependente binária.

A primeira etapa envolveria a análise dos valores NaN, ou seja, nulos. Como apenas a feature "RecommendationType" possuia esses valores e eles representavam aproximadamente 98%, não foi possível excluir as linhas NaN nem substituir por outros valores.

A segunda etapa foi para verificar a quantidade de valores das varíaveis categoricas. As features "DealerId", "SearchId", "deviceid" possuiam inúmeros valores, inviabilizando a possibilidade de criar dummies ou modelo multinível para elas

A terceira etapa foi a de criar as varíaveis dummies para as variáveis categóricas "Distributor", "DealerETAGroup", "DealerOpened", "Position".

```python
df = dataset[['Distributor', 'DealerETAGroup', 'DealerOpened', 'lat', 'long', 'Position', 'Quality', 'Price', 'Conversion']]
df.head(5)
```

# <font color=lightcoral>4) Criando variáveis dummies</font>
***

```python
# Abaixo foram criadas variáveis dummies para as variáveis categoricas

df = pd.get_dummies(df, columns=["Distributor", "DealerETAGroup", "DealerOpened", "Position"])
df.head(5)
```
